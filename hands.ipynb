```python
import os
import glob
import pandas as pd
```

```python
df_hands = []
for filename in glob.iglob(r'C:\Users\avida\Downloads\final_files\extraFiles\Training/**/*.csv',recursive=True):
    
    names= filename.split('/')[-1].split('.')[0].lower()
    
    hands_csv=pd.read_csv(filename) 
    time_reduce=hands_csv.head(1)["Time"][0]
    hands_csv=hands_csv.drop(hands_csv[hands_csv.Time<time_reduce+7].index)
    if 'alone' in names:
        hands_csv['Type']=0
        
    if 'sync' in names:
        hands_csv['Type']=1
        
    if 'spontan' in names:
        hands_csv['Type']=2
        
    df_hands.append(hands_csv)    
    df=pd.concat(df_hands)   
```
